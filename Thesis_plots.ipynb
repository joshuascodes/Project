#### Set the plot styles

```python
from datetime import datetime, timedelta
from matplotlib import rcParams
plt.rc('font', family='serif', serif='Times')
plt.rc('text', usetex=True)
plt.rc('xtick', labelsize=11)
plt.rc('ytick', labelsize=11)
plt.rc('axes', labelsize=11)


sns.set(rc={'figure.figsize':(7.0, 5)})
custom_style = {'figure.facecolor': 'white',
               'axes.labelcolor': 'black',
               'text.color': 'black',
                'xtick.color': 'black',
                'ytick.color': 'black',
               'axes.edgecolor': '.4'}
```

```python
Z = ('-', '-', '-')
sns.set_style()
sns.set_style("whitegrid", custom_style)
sns.set_palette("cubehelix", 1)
plt.xlabel("Year")
plt.ylabel("Q Ratio")

fig = sns.lineplot(data = df3, style = 'Z', dashes = False)
plt.draw()

#plt.savefig('C:\\Users\Joshu\OneDrive\Thesis.2020\Final stretch\images\Q_plotYES.JPEG', dpi=250);
plt.show();
```
